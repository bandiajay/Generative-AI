<h4> Step 3.2 : Initialise the recognizer </h4>

<h4> Step 3.4: Convert to text </h4>

*Duration* is a parameter used to estimate the noise level. Here` 0.5  ` makes recognizer to analyze the ambient noise for half a second

```python
with sr.AudioFile("gettysburg10.wav") as input_audio:

  # Adjust for ambient noise and record the audio
  recognizer.adjust_for_ambient_noise(input_audio, duration=0.5)
  output_text = recognizer.listen(input_audio)
  # Using Google Speech Recognition to convert audio to text
  try:
          print('Converting audio transcripts into text ...')
          text = recognizer.recognize_google(output_text)
          print(text)

  except:
          print('Sorry.. run again...')
```
