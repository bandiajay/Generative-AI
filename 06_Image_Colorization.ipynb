<h4>Step 3.3 : Define an output image</h4>

Once we convert colored image to greyscale, it needs to be stored in an image. So, declare a variable to store the grayscale image.

```python
output_img = r'grayscale.png'
```

<h4>Step 3.4: Remove the color and Save the new Grayscale image into output image </h4>



*   `imread` function read the input image.
*   `cvtColor` function used for converting an image from one color space to another.
      * First argument is input image whose color needs to be changed.
      * second argument specifies the type of color conversion that should take place. Here `cv2.COLOR_BGR2GRAY` indicates that the conversion should be from BGR color space to grayscale.
*   `imwrite` function stores the converted data into image.



```python
# Load the color image
color_image = cv2.imread(input_img)

# Check if the image is loaded successfully
if color_image is None:
    print(f"Error: Unable to load the color image from {input_img}")
else:
    # Convert the color image to grayscale
    gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

    # Save the resulting black and white image
    success = cv2.imwrite(output_img, gray_image)

    if success:
        print(f"Black and white image saved successfully to {output_img}")
    else:
        print(f"Error: Unable to save the black and white image to {output_img}")
```

<h4> Step 3.5: Display converted grayscale image </h4>
