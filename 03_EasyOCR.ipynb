<h3> Step 3 : Write Source Code </h3>

<h4> Step 3.1 : Import easyocr, PIL modules </h4>

Image is a class from PIL library that provides functions to open, save and manipulate images.

```python
import easyocr
from PIL import Image
```

<h4> Step 3.2 : Set the language.</h4>

Here we are setting the language to English. You can follow <a href="https://www.jaided.ai/easyocr/documentation/"> EasyOCR </a> documentation if you want to try other languages.

```python
reader = easyocr.Reader(['en'])
```

<h4> Step 3.3 : Provide an input image </h4>

Here I selected an image from web, saved it as `handwrittentext.jpg` and uploaded it to my google drive. You need to test your own image (by downloading an image with clear text, saving with a good name with extension (.jpg or .png) and uploading to your google drive) in place of `handwrittentext.jpg`

```python
input_image_name = r'handwrittentext.jpg'
input_image = Image.open(input_image_name)
display(input_image)
```

<h4> Step 3.4 : Extract the text</h4>

The <code> readtext </code> method returns a list of tuples, where each tuple represents a detected text instance in the image.
<ul>
<li> index 0 gives Bounding box co-ordinates of text in that image </li>
<li> index 1 gives the text present in the image</li>
<li> index 2 gives confident score which says the accuracy of extracted text </li>
</ul>

```python
result = reader.readtext(input_image)
```

As we are only interested in the text, we will consider only index 1.

```python
for text in result:
    print(text[1])
```
