<b>Introduction: </b> <p> Tesseract is an open-source Optical Character Recognition (OCR) engine that is highly regarded for its accuracy and flexibility. Originally developed by Hewlett-Packard in the 1980s and later sponsored by Google, Tesseract has evolved into one of the most powerful OCR engines available today. It can read a wide range of image formats and convert them into over 100 different languages. Tesseract is particularly praised for its ability to adapt to various fonts and languages, making it a popular choice for developers and researchers in the field of document analysis and archival projects. Its robustness and extensibility make it a go-to solution for anyone needing to extract readable text from digital images efficiently.</p>

<h4> Import <code>pytesseract</code>, <code>PIL</code> modules </h4>

The `Image` module provides a class that includes methods to open, manipulate, and save images.

```python
import pytesseract
from PIL import Image
```

<h4>  Provide an input image </h4>

Here, I selected an image from web, saved it as `test.jpg` and uploaded it to my google drive. You need to test your own image (by downloading an image with clear text, saving with a good name with extension (.jpg or .png) and uploading to your google drive) in place of `test.jpg`

```python
input_image_file = r'test.jpg'
input_image = Image.open(input_image_file)
display(input_image)
```

<h4> Extract the text </h4>

```python
output_text = pytesseract.image_to_string(input_image)
print(output_text)
```
