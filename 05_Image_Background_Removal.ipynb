<h4> Step 3.1 : Import rembg, PIL modules </h4>

The `Image` class from `PIL` package encompasses a comprehensive suite of functions essential for image processing tasks. On the other hand, the `remove` class from `rembg` library offers specialized mechanisms dedicated to background removal from images.

```python
from rembg import remove
from PIL import Image
```

<h4>Step 3.2: Provide an input image </h4>

Here I selected an image from web, saved it as `backgroundcut.jpg` and uploaded it to my google drive. You need to test your own image (by downloading an image with nice background, saving with a good name with extension (.jpg or .png) and uploading to your google drive) in place of `backgroundcut.jpg`

```python
input_img = r'backgroundcut.png'
```

```python
from IPython.display import Image
Image(filename='backgroundcut.png')
```

<h4> Step 3.3 : Remove the background and Save the new image without background </h4>

```python
from PIL import Image
```

```python
try:
    inp = Image.open(input_img)
# remove function that takes the image object inp as an argument and performs background removal
    output = remove(inp)
    output_img = r'output.png'
    output.save(output_img)
    print("Background removed and image saved.")
except FileNotFoundError:
    print("The specified file was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

<h4> Step 3.4: Display the output </h4>
